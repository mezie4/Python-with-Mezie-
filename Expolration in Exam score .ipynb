```python
df1=df.pivot_table(index="gender", columns="race/ethnicity", values="average")
#pivot table of race/ethnicity as columns,gender as index and average as values
```

```python
df1.plot(kind='bar', xlabel='Gender', ylabel='Average', title='Average Score of Student By Race/Ethnicity')
#bar chart of average score of student by race/ethnicity
```

```python
df2=df.pivot_table(index="race/ethnicity", columns="gender", values="math score")
#pivot table of race/ethnicity as index,gender as columns and math score as values
```

```python
df2.plot(kind = 'line', xlabel='Race/Ethnicity', ylabel='Math Score', title='Math score of student by race/ethnicity')
#line chart of the math score of student by race/ethnicity
```

```python
df2.plot(kind = 'bar', xlabel='Race/Ethnicity', ylabel='Math score', title='Math score of student by race/ethnicity')
#bar chart of math score of student by race/ethnicity
```

```python
df3=df1[df1.index.isin(['female'])]
df3
#selecting only the female as index
```

```python
df3=df3.T
df3
#Transpose of the df3 table(changeing from horizontal to vertical)
```

```python
df3.plot(kind = 'bar', ylabel='average score', title='Average Score of Female Student')
#Bar chart of female average score
```

```python
df3.plot(kind ='pie', y='female', title='Average Score of Female Student') 
#pie chart of female average score
```
